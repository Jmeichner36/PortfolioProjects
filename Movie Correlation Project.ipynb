```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)


# Read in data

df = pd.read_csv('D:\movies.csv')

```

```python
# Looking at data

df.head()
```

```python
# Searching for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Removing missing data

df = df.dropna()
```

```python
# Column data types

df.dtypes
```

```python
# Change budget, gross, and votes columns to int types

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

```python
pd.set_option('display.max_rows', None)

df.head()
```

```python
# Correcting mistakes between year and released year

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df.head()
```

```python
# Sorting by gross revenue

df = df.sort_values(by=['gross'], inplace=False, ascending=False)

df.head()
```

```python
# Building scatter plot with budget vs. gross


plt.scatter(x=df['budget'], y=df['gross'])

plt.xlabel('Budget')
plt.ylabel('Gross Revenue')
plt.title('Budget vs. Gross Revenue')

plt.show()
```

```python
df.head()
```

```python
# Plotting budget vs. gross with seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
# Pearson correlation

df.corr()
```

```python
# High correlation between budget and gross
# Visualizing correlation matrix

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.title('Correlation Matrix for Numerical Data')
plt.show()
```

```python
# Changing columns with object data types into categories

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
# Correlatin matrix with categories included

correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.title('Correlation Matrix for Numerical Data')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending=False)

sorted_pairs
```
